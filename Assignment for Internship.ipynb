# Task - 6

Business wants to check the Claim Amount for deviation for each transaction, they would like you to calculate as follow

CLAIM_DEVIATION = AVG_CLAIM_AMOUNT_FOR_LAST_30DAYS (same insurance type) / CURRENT_CLAIM_AMOUNT

If the value < 0.5 THEN CLAIM_DEVIATION = 1 ELSE 0 If there is less than 30 days of transaction history THEN

-1 Note: LAST_30DAYS does not include current day

```python
# Fetch max value for TXN_DATE_TIME feature
view.TXN_DATE_TIME.max()
```

```python
# Converting TXN_DATE_TIME feature max value into a datetime form
todays_date = pd.to_datetime('2021-06-30')
todays_date
```

```python
# Look through the values of feature 'TXN_DATE-TIME' and converting it into a datetime form
view['TXN_DATE_TIME'] = pd.to_datetime(view['TXN_DATE_TIME'])
view['TXN_DATE_TIME']
```

```python
# Fetch last 30 days value from feature 'TXN_DATE_TIME'
last_30 = view.loc[(view['TXN_DATE_TIME'] > '2021-05-30') & (view['TXN_DATE_TIME'] < '2021-06-30')] 
last_30
```

```python
# Fetch average claim amount from last 30 days by grouping with INSURANCE_TYPE and converting it into a dictionary
avg_claim = last_30.groupby('INSURANCE_TYPE').mean()['CLAIM_AMOUNT'].to_dict()
avg_claim
```

#### Conclusion Drawn:-
        ~ From the above dataframe we get to know that Life insurance type has the highest average claim amount whereas     Mobile insurance type has least average claim amount.

```python
# define a function for calculating claim_deviation
def Claim_deviation(row):
    if (todays_date - row['TXN_DATE_TIME'])<pd.Timedelta(days=30):
        return -1
    claim_deviation = avg_claim[row['INSURANCE_TYPE']] / row['CLAIM_AMOUNT']
    if claim_deviation < 0.5:
        return 1
    else:
        return 0
```

```python
# Applying the function to the dataframe
view['CLAIM_DEVIATION'] = view.apply(Claim_deviation,axis=1)
view['CLAIM_DEVIATION']
```

```python
# Fetch values for feature CLAIM_DEVIATION
view['CLAIM_DEVIATION'].value_counts()
```

#### Conclusion Drawn:-
    ~ For values less than 0.5 there are 9216 claims whereas 784 claims with transaction history of less than 30 days.

# Task - 8

Mobile & Travel Insurance premium are discounted by 10%

Health and Property Insurance premium are increased by 7%

Life and Motor Insurance premium are marginally increased by 2%

What will be overall change in % of the Premium Amount Collected for all these Customer?

```python
# Fetch the premium value collected
premium_collected = view['PREMIUM_AMOUNT'].sum()
premium_collected
```

```python
# Mobile & Travel premium discounted by 10%
premium_type['Mobile']*=0.9
premium_type['Travel']*=0.9
# Health and Property premium increases by 7%
premium_type['Health']*=1.07
premium_type['Property']*=1.07
# Life and Motor premium increased by 2%
premium_type['Life']*=1.02
premium_type['Motor']*=1.02
```

```python
# Premium after discount and increase
new_premium = sum(premium_type.values())
new_premium
```

```python
# Calculate average premium collected 
(new_premium-premium_collected)*100/premium_collected
```

```python
# Fetch the value of premium collected and value of new premium collected
premium_collected,new_premium
```

#### Conclusion Drawn:-

Business wants to check Claim Velocity which is defined as follow

CLAIM_VELOCITY = NO_OF_CLAIMS_IN_LAST30DAYS (for the current insurance type) / NO_OF_CLAIMS_IN_LAST3DAYS (for the current insurance type)

Note: LAST30DAYS & LAST3DAYS does not include current day

```python
# Fetch current date value
todays_date
```

```python
# Fetch number of values in feature 'CLAIM_STATUS'
view['CLAIM_STATUS'].value_counts()
```

# We had defined last 30 days data in task 6 earlier, similarly defining for last 3 days.

```python
# Fetch values for last 3 days and converting it into timedelta
# Timedelta represents a duration, the difference between two dates or times.
last_3_days = view.loc[(view['TXN_DATE_TIME'] > todays_date-pd.Timedelta(days=4)) & (view['TXN_DATE_TIME'] < todays_date)]
last_3_days
```

```python
# Calculating total approved claims insurance type wise in last 30 days
claims_in_last_30_days = pd.DataFrame.from_dict(last_30[last_30['CLAIM_STATUS']=='A'].groupby('INSURANCE_TYPE').size().to_dict(),orient='index',columns=['NO_OF_CLAIMS_IN_LAST30DAYS'])
claims_in_last_30_days
```

```python
# Calculating total approved claims insurance type wise in last 3 days
claims_in_last_3_days = pd.DataFrame.from_dict(last_3_days[last_3_days['CLAIM_STATUS']=='A'].groupby('INSURANCE_TYPE').size().to_dict(),orient='index',columns=['NO_OF_CLAIMS_IN_LAST3DAYS'])
claims_in_last_3_days
```

```python
# Fetch the value of approved claims by joining claims in last 30 and 3 days
approved_claims = claims_in_last_30_days.join(claims_in_last_3_days)
approved_claims
```

```python
# Create new feature CLAIM_VELOCITY by calculating difference between no. of claims in last 30 days and 3 days
approved_claims['CLAIM_VELOCITY'] = approved_claims['NO_OF_CLAIMS_IN_LAST30DAYS']/approved_claims['NO_OF_CLAIMS_IN_LAST3DAYS']
approved_claims['CLAIM_VELOCITY']
```

#### Conclusion Drawn:-
    ~From the above table we get to know that the CLAIM_VELOCITY for Mobile is very high and least CLAIM_VELOCITY of Travel.
